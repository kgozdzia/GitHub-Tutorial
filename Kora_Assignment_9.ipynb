```python
from math import pi
#Homework 9A use Turtle to draw basic geometric shapes (triangle, square, hexagon, circle). 
#Create functions for each rather than inputting the instructions over and over again.
#Create a basic drawing using one or more of these basic shapes.

#import TurtlePlus
from ColabTurtlePlus.Turtle import *
#use clearscreen() to re-run commands in the cell
clearscreen()
# set screen size
setup(800,800)

#create square function
def drawSquare(turtleObj, sideLength):
  for i in range(4):
    turtleObj.fd(sideLength)
    turtleObj.lt(90)

#create polygon function
def drawPolygon(turtleObj, sideNumber, sideLength):
  for side in range(sideNumber):
    turtleObj.fd(sideLength)
    turtleObj.lt(360/sideNumber)

#create circle function
def drawCircle (turtleObj, radius):
  from math import pi as pi
  circumference = 2*pi*radius
  sides = int(circumference/3)+1
  length = circumference/sides
  drawPolygon(turtleObj, sides, length)

#create arc function
def drawArc(turtleObj, radius, startAngle):
  from math import pi as pi                 
  arcLength = 2*pi*radius*startAngle/360
  sides = int(arcLength/3)+1
  stepLength = arcLength/sides 
  stepAngle = startAngle/sides

  for i in range(sides):
    turtleObj.fd(stepLength)
    turtleObj.lt(stepAngle)

#create reverse arc function
def drawArcReverse(turtleObj, radius, startAngle):
  from math import pi as pi                 
  arcLength = 2*pi*radius*startAngle/360
  sides = int(arcLength/3)+1
  stepLength = arcLength/sides 
  stepAngle = startAngle/sides

  for i in range(sides):
    turtleObj.fd(stepLength)
    turtleObj.rt(stepAngle)
#create equilateral triangle function
def drawEqualTri(turtleObj, sideLength):
  for i in range(3):
   turtleObj.forward(sideLength)
   turtleObj.left(360/3)

# create turtle objects
turtleA = Turtle()
turtleA.color(255, 0, 0)
turtleB = Turtle()
turtleB.color(255, 128, 0)
turtleC = Turtle()
turtleC.color((240, 240, 0))
turtleD = Turtle()
turtleD.color((25,240,15))
turtleE = Turtle()
turtleE.color((20,40,255))
turtleF = Turtle()
turtleF.color((180, 20, 170))
# create basic drawing
turtleA.jumpto(-380,200)
drawEqualTri(turtleA, 200)
turtleB.jumpto(-240, 120)
drawSquare(turtleB, 120)
turtleC.jumpto(-125, 65)
drawPolygon(turtleC, 5, 70)
turtleD.jumpto(-57, 27)
drawPolygon(turtleD, 6, 40)
turtleE.jumpto(-11, 0)
drawPolygon(turtleE, 7, 25)
turtleF.jumpto(20, -22)
drawPolygon(turtleF, 8, 15)
```

```python
#Homework 9B - use the function(s) you created for Homework_9A, as well as any other you need, and create a complex drawing using one or more turtles.
clearscreen()
# set screen size
setup(1600,800)
#set screen color
bgcolor(20, 20, 40)

# create turtle objects
starObj = Turtle()
starObj.color((240, 240, 185))
buildingObj = Turtle()
buildingObj.color((10, 10, 20))
windowObj = Turtle()
windowObj.color(240, 230, 120)
moonObj = Turtle()
moonObj.color(235, 250, 185)

# create star outline function
def drawFilledStar(turtleObj, sideLength):
  turtleObj.begin_fill()
  for i in range(10):
    result = i%2
    if result == 0:
      turtleObj.rt(144)
    else:
      turtleObj.lt(72)
    turtleObj.fd(sideLength)
  turtleObj.end_fill()

#create random star function
def randomStar(turtleObj):
  #create random generator for size and location
  from random import randint as randomInt
  for i in range (30):
    randLength = randomInt(2, 20)
    randXVal = randomInt(-800, 800)
    randYVal = randomInt(-370, 400)
    turtleObj.jumpto(randXVal, randYVal)
    drawFilledStar(turtleObj, randLength)

# create moon function
def drawCrescent(turtleObj, radius):
  moonObj.lt(160)
  moonObj.begin_fill()
  drawArc(moonObj, radius, 240)
  moonObj.rt(200)
  drawArcReverse(moonObj, .885*radius, 202)
  moonObj.end_fill()

# create building function
def drawSkyline(turtleObj):
  buildingObj.jumpto(-800, -400)
  buildingObj.begin_fill()
  buildingObj.lt(90)
  buildingObj.fd(180)
  buildingObj.rt(90)
  buildingObj.fd(200)
  buildingObj.lt(90)
  buildingObj.fd(220)
  buildingObj.rt(90)
  buildingObj.fd(120)
  buildingObj.rt(90)
  buildingObj.fd(350)
  buildingObj.lt(90)
  buildingObj.fd(100)
  buildingObj.lt(90)
  buildingObj.fd(620)
  buildingObj.rt(90)
  buildingObj.fd(250)
  buildingObj.rt(90)
  buildingObj.fd(200)
  buildingObj.lt(90)
  buildingObj.fd(270)
  buildingObj.rt(90)
  buildingObj.fd(350)
  buildingObj.lt(90)
  buildingObj.fd(200)
  buildingObj.lt(90)
  buildingObj.fd(500)
  buildingObj.rt(90)
  buildingObj.fd(180)
  buildingObj.rt(90)
  buildingObj.fd(600)
  buildingObj.lt(90)
  buildingObj.fd(150)
  buildingObj.lt(90)
  buildingObj.fd(250)
  buildingObj.rt(90)
  buildingObj.fd(350)
  buildingObj.rt(90)
  buildingObj.fd(400)
  buildingObj.rt(90)
  buildingObj.fd(1600)
  buildingObj.end_fill()

# create windows(?)
def drawWindows(turtleObj, height, width, gapX, gapY,perRow, rows):
  for i in range(rows):
    for i in range(perRow):
      turtleObj.begin_fill()
      for i in range (2):
        turtleObj.fd(width)
        turtleObj.rt(90)
        turtleObj.fd(height)
        turtleObj.rt(90)
      turtleObj.end_fill()
      turtleObj.penup()
      turtleObj.fd(width+gapX)
      turtleObj.pendown()
    turtleObj.penup()
    turtleObj.bk((perRow*width)+(perRow*gapX))
    turtleObj.rt(90)
    turtleObj.fd(height+gapY)
    turtleObj.lt(90)
    turtleObj.pendown()

randomStar(starObj)

moonObj.jumpto(210, 300)
drawCrescent(moonObj, 100)

drawSkyline(buildingObj)

windowObj.jumpto(-582, -20)
drawWindows(windowObj, 70, 17, 17, 70, 3, 3)

windowObj.jumpto(-360, 220)
drawWindows(windowObj, 50, 80, 45, 80, 2, 5)

windowObj.jumpto(360, 195)
drawWindows(windowObj, 120, 20, 20, 30, 4, 4)

windowObj.jumpto(690, -150)
drawWindows(windowObj, 40, 40, 10, 10, 3, 3)

#capture the data written to the screen
screen_obj = Screen()
#save screen as svg
screen_obj.saveSVG('Cityscape.svg', False)
```
